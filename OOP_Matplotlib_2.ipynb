```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,2*np.pi,50)

sin_x = np.sin(x)
```

```python
fig, ax = plt.subplots(figsize=[7,3])

ax.plot(x, sin_x, marker='*', markersize=3, linestyle='');
# ax.set_aspect('equal');
```

### Changing the Plot's Color

```python
fig, ax = plt.subplots(figsize=[10,5])
ax.set_aspect('equal')

ax.plot(x,sin_x,
        marker='*',
        markersize=8,
        color='red',
        markerfacecolor='blue',
        markeredgecolor='yellow',
        linestyle='-.');
```

**The Color Attribute in any matplotlib plot changes the color of the main thing being plotted.In the above caes, that would be the line plot**
